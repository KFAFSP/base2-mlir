```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import struct
from sklearn.metrics import mean_squared_error
```

```python
# function to read data from std kernel result
def read_result(file):
    df = pd.read_csv(file)
    data_std = df.loc[:, "c_std"]
    array_std = data_std.to_numpy(dtype=float)
    data_sf = df.loc[:, "c_sf"]
    array_sf = data_sf.to_numpy(dtype=float)
    return array_std, array_sf
```

```python
# function to calculate MSEs of given filenames
def mse(file_str):
    v_std, v_sf = read_result(file_str)
    v_mse = mean_squared_error(v_std, v_sf)
    # TODO: use numpy maybe
    return v_mse
```

```python
# initialize arrays to store MSEs
v_mse_exp = np.zeros(6)
v_mse_frac = np.zeros(10)

# fill the MSE arrays
# 6_52
file_str = "./result/result_6_52_frac.csv"
v_mse_exp[5] = mse(file_str)
# 7_52
file_str = "./result/result_7_52_frac.csv"
v_mse_exp[4] = mse(file_str)
# 8_52
file_str = "./result/result_8_52_frac.csv"
v_mse_exp[3] = mse(file_str)
# 9_52
file_str = "./result/result_9_52_frac.csv"
v_mse_exp[2] = mse(file_str)
# 10_52
file_str = "./result/result_10_52_frac.csv"
v_mse_exp[1] = mse(file_str)
# 11_52
file_str = "./result/result_11_52_frac.csv"
v_mse_exp[0] = mse(file_str)
v_mse_frac[0] = mse(file_str)
# 11_45
file_str = "./result/result_11_45_frac.csv"
v_mse_frac[1] = mse(file_str)
# 11_40
file_str = "./result/result_11_40_frac.csv"
v_mse_frac[2] = mse(file_str)
# 11_35
file_str = "./result/result_11_35_frac.csv"
v_mse_frac[3] = mse(file_str)
# 11_30
file_str = "./result/result_11_30_frac.csv"
v_mse_frac[4] = mse(file_str)
# 11_29
file_str = "./result/result_11_29_frac.csv"
v_mse_frac[5] = mse(file_str)
# 11_28
file_str = "./result/result_11_28_frac.csv"
v_mse_frac[6] = mse(file_str)
# 11_27
file_str = "./result/result_11_27_frac.csv"
v_mse_frac[7] = mse(file_str)
# 11_26
file_str = "./result/result_11_26_frac.csv"
v_mse_frac[8] = mse(file_str)
# 11_25
file_str = "./result/result_11_25_frac.csv"
v_mse_frac[9] = mse(file_str)
```

```python
# plot the MSEs
exp_bits = [6, 7, 8, 9, 10, 11]
frac_bits= [52, 45, 40, 35, 30, 29, 28, 27, 26, 25]
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle("Random numbers with small exponent")
ax1.plot(exp_bits, v_mse_exp, marker='.') ## v is the output
ax1.set_xlabel("exponent bits with 52-bit frac")
ax1.set_ylabel("Mean Square Errors")
ax2.plot(frac_bits, v_mse_frac, marker='.')
ax2.set_xlabel("fractional bits with 11-bit exp")
plt.show()
```
