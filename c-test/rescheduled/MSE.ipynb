```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import struct
from sklearn.metrics import mean_squared_error
```

```python
# function to read data from std kernel result
def read_std(file):
    df = pd.read_csv(file)
    fv_data = df.loc[:, "v"]
    fv_array = fv_data.to_numpy(dtype=float)
    ft_data = df.loc[:, "t"]
    ft_array = ft_data.to_numpy(dtype=float)
    fr_data = df.loc[:, "r"]
    fr_array = fr_data.to_numpy(dtype=float)
    ft0_data = df.loc[:, "t0"]
    ft0_array = ft0_data.to_numpy(dtype=float)
    ft1_data = df.loc[:, "t1"]
    ft1_array = ft1_data.to_numpy(dtype=float)
    ft2_data = df.loc[:, "t2"]
    ft2_array = ft2_data.to_numpy(dtype=float)
    ft3_data = df.loc[:, "t3"]
    ft3_array = ft3_data.to_numpy(dtype=float)
    return fv_array, ft_array, fr_array, ft0_array, ft1_array, ft2_array, ft3_array
```

```python
# function to convert 64-bit hex numbers into double
def to_float(a):
    for i in range(len(a)):
        a[i] = int(a[i], base=16)
        byte_str = struct.pack('>Q', a[i])
        a[i] = struct.unpack('>d', byte_str)[0]
```

```python
# function to read data from sf kernel result
def read_sf(file):
    df = pd.read_csv(file)
    iv_data = df.loc[:, "v"]
    iv_array = iv_data.to_numpy()
    to_float(iv_array)
    it_data = df.loc[:, "t"]
    it_array = it_data.to_numpy()
    to_float(it_array)
    ir_data = df.loc[:, "r"]
    ir_array = ir_data.to_numpy()
    to_float(ir_array)
    it0_data = df.loc[:, "t0"]
    it0_array = it0_data.to_numpy()
    to_float(it0_array)
    it1_data = df.loc[:, "t1"]
    it1_array = it1_data.to_numpy()
    to_float(it1_array)
    it2_data = df.loc[:, "t2"]
    it2_array = it2_data.to_numpy()
    to_float(it2_array)
    it3_data = df.loc[:, "t3"]
    it3_array = it3_data.to_numpy()
    to_float(it3_array)
    return iv_array, it_array, ir_array, it0_array, it1_array, it2_array, it3_array
```

```python
# function to calculate MSEs of given filenames
def mse(file_std, file_sf):
    fv, ft, fr, ft0, ft1, ft2, ft3 = read_std(file_std)
    iv, it, ir, it0, it1, it2, it3 = read_sf(file_sf)
    v_mse = mean_squared_error(fv, iv)
    t_mse = mean_squared_error(ft, it)
    r_mse = mean_squared_error(fr, ir)
    t0_mse = mean_squared_error(ft0, it0)
    t1_mse = mean_squared_error(ft1, it1)
    t2_mse = mean_squared_error(ft2, it2)
    t3_mse = mean_squared_error(ft3, it3) ## TODO: use numpy maybe
    return v_mse, t_mse, r_mse, t0_mse, t1_mse, t2_mse, t3_mse
```

```python
# initialize arrays to store MSEs
v_mse = np.zeros(10)
t_mse = np.zeros(10)
r_mse = np.zeros(10)
t0_mse = np.zeros(10)
t1_mse = np.zeros(10)
t2_mse = np.zeros(10)
t3_mse = np.zeros(10) ## TODO: ten different test with different parameters
```
